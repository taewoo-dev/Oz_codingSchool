```python
# ê¸°ë³¸ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¼ íŒŒì´ì¬ ì„¤ì¹˜ì‹œ ìë™ìœ¼ë¡œ ì„¤ì¹˜ë©ë‹ˆë‹¤!
# random ëª¨ë“ˆì€ ëœë¤í•œ ìˆ«ìë¥¼ ìƒì„±í•  ë•Œ ì‚¬ìš©í•©ë‹ˆë‹¤.
import random
import string

ALPHABET = string.ascii_lowercase

# ê²Œì„ì— ì‚¬ìš©ë  ë‹¨ì–´ ëª©ë¡
words = ["apple", "banana", "orange", "grape", "lemon"]


# í–‰ë§¨ ê·¸ë¦¼
hangman_pics = [
    """
     ------
     |    |
     |
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |    |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\
     |   /
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\
     |   / \
     |
    ---""",
]


class HangmanGame:

    def __init__(self) -> None:

        self.words = []
        self.hangman_pics = []
        self.hangman_word = ''
        self.hidden_word = []
        self.life = 0
        self.count = 0
        self.answer_word = []

    def start_setting(self,words : list,hangman_pics : list) -> None:

        self.words = words
        self.hangman_pics = hangman_pics
        self.life = len(self.hangman_pics)
        self.hangman_word = random.choice(words) 
        self.hidden_word = len(self.hangman_word)*['_']

        self.hangman_pics.reverse()

    def display(self) -> None:

        print(" ".join(self.hidden_word))
        print(self.hangman_pics[self.life-1])

    def start_ment(self) -> None:

        print("hangman gameì„ ì‹œì‘í•©ë‹ˆë‹¤")

        print(f"ê²Œì„ ì‹œì‘ ì‹œ ëª©ìˆ¨ì€ {self.life}ê°œì´ë©°, ì´ 10íšŒì˜ ì •ë‹µì„ ë§ì¶œ ê¸°íšŒê°€ ì£¼ì–´ì§‘ë‹ˆë‹¤")


    def user_input_check(self,user_input : str) -> bool:
        
        if len(user_input) != 1:
            print("ì•ŒíŒŒë²³ì´ ì—¬ëŸ¬ê°œ ì…ë ¥ë˜ì—ˆìŠµë‹ˆë‹¤. ë‹¤ì‹œ ì‹œë„í•´ì£¼ì„¸ìš”")
            return True
        elif user_input not in ALPHABET:
            print("ì•ŒíŒŒë²³ì´ ì•„ë‹Œ ê¸€ìê°€ ì…ë ¥ë˜ì—ˆìŠµë‹ˆë‹¤. ë‹¤ì‹œ ì‹œë„í•´ ì£¼ì„¸ìš”")
            return True 

        return False
    
    def uncover_text(self,user_input : str) -> None:
        for index, value in enumerate(self.hangman_word):
            if value == user_input:
                    self.hidden_word[index] = user_input

    def view_life(self) -> None:

        print(f"í˜„ì¬ ë‚¨ì€ ëª©ìˆ¨ì€ [{self.life}/7] ì…ë‹ˆë‹¤")

        print(f"í˜„ì¬ ì‹œë„ íšŸìˆ˜ëŠ” [{self.count}/10] ì…ë‹ˆë‹¤")

        print(f"í˜„ì¬ í™•ì¸í•œ ì•ŒíŒŒë²³ì€ [{self.answer_word}] ì…ë‹ˆë‹¤")

    def check_duplicates(self,user_input : str) -> bool:

        if user_input in self.answer_word:
            print("ì´ë¯¸ í™•ì¸í•œ ì•ŒíŒŒë²³ì…ë‹ˆë‹¤, ë‹¤ì‹œ ì‹œë„í•´ì£¼ì„¸ìš”")
            return True
        else:
            self.answer_word.append(user_input)
            return False


    
    def play(self) -> None:
        
        self.start_ment()

        self.display()

        while True:
            
            try:

                user_input = input("ìƒê°í•˜ëŠ” ì•ŒíŒŒë²³ ë‹¨ì–´ í•˜ë‚˜ë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”!, ê²Œì„ì˜ ì¢…ë£Œë¥¼ ì›í•˜ì‹œë©´ 'ì¢…ë£Œ'ë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”").strip().lower()

                if user_input == "ì¢…ë£Œ":

                    print("ê²Œì„ì„ ì¢…ë£Œí•©ë‹ˆë‹¤")
                    break

                if self.user_input_check(user_input):
                    continue

                if self.check_duplicates(user_input):
                    continue

                if user_input in self.hangman_word:

                    print(f"{user_input}ì€ ë§ìŠµë‹ˆë‹¤!")

                    self.uncover_text(user_input)

                    self.count += 1

                    self.view_life()

                    self.display()
                    

                    if "".join(self.hidden_word) == self.hangman_word:
                        print("ì •ë‹µì„ ë§ì¶”ì…¨ìŠµë‹ˆë‹¤~!~!~! ê²Œì„ì„ ì¢…ë£Œí•©ë‹ˆë‹¤")
                        break

                else:
                    self.life -= 1

                    self.count += 1

                    print("ë‹¹ì‹ ì€ í‹€ë ¸ìŠµë‹ˆë‹¤...")

                    self.view_life()

                    self.display()
                    
                    if self.life == 0 or self.count == 10:

                        print("ë‹¹ì‹ ì´ ì‚¬ëŒì„ ì£½ì˜€ìŠµë‹ˆë‹¤...")

                        print("ê²Œì„ì„ ì¢…ë£Œí•©ë‹ˆë‹¤")

                        break
            except:
                print("ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤. ê°’ì„ ë‹¤ì‹œ ì…ë ¥í•´ ì£¼ì„¸ìš”")

if __name__ == "__main__":
    game = HangmanGame()
    game.start_setting(words,hangman_pics)
    game.play()
```
