[과제 8: 메인 함수 main]

목표: main 함수를 작성하여 프로그램의 주요 로직을 실행합니다.

설명:
- BungeoppangShop 객체를 생성합니다.
- 사용자에게 모드를 선택하도록 요청합니다.
- "주문" 모드에서는 손님의 주문을 처리하고, "관리자" 모드에서는 재고를 추가할 수 있습니다.
- 프로그램이 종료될 때 총 매출을 계산하고 데이터를 저장합니다.


```python
class BungeoppangShop:
    def __init__(self) -> None:
        
        bungeoppang_type = ["팥붕어빵","슈크림붕어빵","초코붕어빵"]
        bungeoppang_stock = [10,8,5]
        bungeoppang_prices = [1000,1200,1500]
        bungeoppang_sales = [0,0,0]
        
        self.stock : dict[str,int] = dict(zip(bungeoppang_type,bungeoppang_stock))
        self.prices : dict[str,int] = dict(zip(bungeoppang_type,bungeoppang_prices))
        self.sales : dict[str,int] = dict(zip(bungeoppang_type,bungeoppang_sales))

    def check_stock(self) -> None:
        print(f"남은 붕어빵의 재고 : {self.stock}")

    def process_order(self, bread_type : str, bread_count : int ) -> None:
        
        if self.stock[bread_type] < bread_count or self.stock[bread_type]==0:
            print(f"{bread_type}는 재고가 부족합니다. 다른 붕어빵을 골라주세요")
        else:
            print(f"{bread_type}은 재고가 충분합니다")
            self.stock[bread_type] -= bread_count
            self.sales[bread_type] += bread_count
            print(f"{bread_type}을 {bread_count}개 판매합니다")
            self.check_stock()
    
    def admin_mode(self) -> None:
        
        try:
            bread_type, additional_stock = input("추가할 붕어빵의 종류와 개수를 입력해주세요 : ex) 팥붕어빵, 10개").strip().split(",")
        except:
            print("값을 잘못입력하셨습니다. 다시 시도해주세요")
        else:
            print(f"{bread_type}을 {additional_stock}개 추가하였습니다")
            self.stock[bread_type] += int(additional_stock)
            self.check_stock()
    
    def calculate_total_sales(self) -> None:

        totla_sales = 0

        for i,j,k in zip(self.sales.keys(),self.sales.values(),self.prices.values()):
            totla_sales += j*k
            print(f"{i}의 판매량은 {j}개로 판매 금액은 {j*k}원 입니다")

        print(f"붕어빵 총 판매 금액은 {totla_sales}원 입니다")

    def check_out_of_stock(self) -> bool:
        if not sum(self.stock.values()):
            return True
        else:
            return False
        
def main():

    oz = BungeoppangShop()

    while True:
        
        mode = input("원하시는 모드를 입력하세요. 1.'주문' 2.'관리자' 프로그램의 종료를 원하시면 '종료'를 입력하세요.").strip()

        if mode == "주문":
            print("주문모드를 실행합니다")

            while True:
                try:
                    if oz.check_out_of_stock():
                        print("붕어빵이 모두 소진되어 판매를 중단합니다.")
                        break

                    type = input("주문하실 붕어빵의 종류를 입력해주세요 : ex) 팥붕어빵,슈크림붕어빵,초코붕어빵 주문을 마치시면 '종료'를 입력해주세요").strip()
                    
                    if type == "종료":
                        print("주문모드를 종료합니다")
                        break
                    count = int(input(f"주문하실 {type}의 갯수를 입력해주세요 : "))
                    oz.process_order(type,count)

                except:
                    print("잘못된 값을 입력하셨습니다. 주문을 다시 해주세요")

        elif mode == "관리자":
            print("관리자모드를 실행합니다")

            while True:
                try:
                    oz.admin_mode()

                    admin_check = input("재고를 더 추가하시겠습니까? 원하지 않으시면 '종료'를 입력해주세요")
                    if admin_check == "종료":
                        print("관리자모드를 종료합니다")
                        break

                except:
                    print("잘못된 값을 입력하셨습니다. 재고 추가 여부를 다시 입력해주세요")
                
        
        elif mode == "종료":
            oz.calculate_total_sales()
            break
        else:
            print("잘못된 값을 입력하셨습니다. 다시 시도해주세요")

if __name__ == "__main__":
    main()
```
